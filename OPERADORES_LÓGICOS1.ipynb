```python
print(f"--- COMPARADORES ---")

setpoint = 25.0
leitura = 25.0
limite_segurança = 30.0

print(f"--- Diagnóstico do Sensor ({leitura}ºC) ---")

# Igualdade (==) e Diferença (!=)
print(f"alvo atingido? {leitura == setpoint}")
print(f"está fora do alvo? {leitura !=setpoint}")

# Maior (>) e Menor (<)
print(f"perigo acima de 30? {leitura > limite_segurança}")
print(f"está frio (abaixo de 20)? {leitura < 20.0}")

# Maior ou Igual (>=) e Menor ou Igual (<=)
print(f"ligar ventoinha (>=25)? {leitura>=25.0}")
print(f"dentro da margem segura (<=30)? {leitura <= limite_segurança}")

print(f"--------------------- OPERADORES LOGICOS ------------------")

botao_start = True # Botão pressionado
porta_fechada = True # Sensor de porta
emergencia_ativa = False # Botão de emergência solto
nivel_oleo = 15 # Nível baixo (minimo 20)
print("\n--- Verificação de Segurança ---")
# AND: Rigoroso. Tudo tem que estar certo.
pronta_para_ligar = botao_start and porta_fechada
print(f"Pode iniciar motor? {pronta_para_ligar}")
# OR: Flexível. Qualquer erro dispara o alarme.
alerta_geral = (nivel_oleo < 20) or emergencia_ativa
print(f"Painel de Alerta aceso? {alerta_geral}")
# NOT: Inversão.
sistema_seguro = not alerta_geral
print(f"Sistema seguro para operar? {sistema_seguro}")

```
