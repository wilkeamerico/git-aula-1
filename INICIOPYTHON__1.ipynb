```python
# Lista de IDs autorizados (Whitelist)
admins = [101, 202, 999]
id_lido = int(input("Digite seu ID: "))
if id_lido in admins:
  print("Bem-vindo, Administrador.")
else:
  print("Acesso comum.")
# Verificando comandos proibidos (Blacklist)
comandos_proibidos = ["FORMAT", "RESET"]
cmd = input("Comando: ")
if cmd not in comandos_proibidos:
  print(f"Executando: {cmd}")
else:
  print("⛔ AÇÃO BLOQUEADA!")
```
