```python
import math
# Exemplo: Linearização de NTC (Equação Steinhart-Hart simplificada)
resistencia = 10000.0
beta = 3950.0
# T = 1 / ( (1/T0) + (1/B) * ln(R/R0) )
temp_k = 1 / ( (1/298.15) + (1/beta) * math.log(resistencia / 10000.0) )
print(f"Temperatura: {temp_k - 273.15:.2f}ºC")
# Exemplo: Seno de 90 graus
seno = math.sin(math.radians(90))
print(f"Seno de 90º: {seno}")
```

```python
import time
import random
# ==========================================
# 1. CONFIGURAÇÕES E CONSTANTES
# ==========================================
print("--- Inicializando Firmware v1.0 ---")
# Configurações do Sistema (Dicionário - Editável)
config = {
"id_estacao": "ESTACAO_SOLAR_01",
"intervalo_leitura": 2,
# Segundos
"tamanho_buffer": 5,
# Quantas leituras para média móvel
"limite_temp_alerta": 40.0  # Graus Celsius
}
# Definição de Hardware (Tupla - Imutável)
# Exemplo: (Pino Sensor, Pino Alerta)
PINOS_ESP32 = (34, 2)
# Variável Global para Média Móvel (Lista)
buffer_irradiancia = []

# ==========================================
# 2. FUNÇÕES AUXILIARES (MODULARIZAÇÃO)
# ==========================================

def simular_leitura_sensor():
    """
    Simula o hardware enviando uma String suja via Serial.
    Retorna algo como: "  IRRAD:850;TEMP:32.5\n  " ou None se falhar.
    """
    # Simula desconexão aleatória (10% de chance)
    if random.random() < 0.1:
        return None

    # Gera valores aleatórios
    irr = random.randint(200, 1000)
    temp = random.uniform(25.0, 45.0)

    # Monta a string suja (Simulando protocolo Serial)
    dado_bruto = f"  IRRAD:{irr};TEMP:{temp:.2f}\n  "
    return dado_bruto


def tratar_dados(dado_bruto):
    """
    Recebe a string bruta, limpa e extrai os números.
    Usa: Strip, Split e tratamento de erros de conversão.
    """
    try:
        # 1. Limpeza (Strip) - Remove espaços e quebra de linha
        limpo = dado_bruto.strip()

        # 2. Separação (Split) - Quebra no ponto e vírgula
        # Ex: "IRRAD:800;TEMP:30.0" -> ['IRRAD:800', 'TEMP:30.0']
        partes = limpo.split(";")

        # 3. Extração (Split no :)
        # Pega a parte depois dos dois pontos e converte para float
        valor_irr = float(partes[0].split(":")[1])
        valor_temp = float(partes[1].split(":")[1])

        # Retorna um Dicionário organizado
        return {"irradiancia": valor_irr, "temp": valor_temp}

    except Exception as e:
        print(f"❌ Erro ao processar string: {e}")
        return None


def processar_media(novo_valor):
    """
    Gerencia o buffer circular e calcula a média.
    """
    buffer_irradiancia.append(novo_valor)

    # Se o buffer encher, remove o mais antigo (FIFO)
    if len(buffer_irradiancia) > config["tamanho_buffer"]:
        buffer_irradiancia.pop(0)

    # Calcula média
    return sum(buffer_irradiancia) / len(buffer_irradiancia)


def verificar_alerta(temperatura):
    """
    Verifica se a temperatura excedeu o limite seguro.
    """
    if temperatura > config["limite_temp_alerta"]:
        return True
    return False


# ==========================================
# 3. LOOP PRINCIPAL (MAIN)
# ==========================================
print(f"📡 Sistema {config['id_estacao']} Online.")
print("Pressione STOP para encerrar.")

while True:
    try:
        # --- PASSO A: Leitura ---
        raw = simular_leitura_sensor()

        if raw is None:
            # Levanta um erro controlado para testar o sistema de reconexão
            raise OSError("Sem resposta do sensor")

        # --- PASSO B: Processamento ---
        dados = tratar_dados(raw)

        if dados:
            irr = dados["irradiancia"]
            temp = dados["temp"]

            # --- PASSO C: Inteligência ---
            media_solar = processar_media(irr)
            alerta = verificar_alerta(temp)

            # --- PASSO D: Saída/Atuação ---
            print("-" * 30)
            print(f"☀️ Instantâneo: {irr} W/m²")
            print(f"📊 Média (5s):  {media_solar:.1f} W/m²")
            print(f"🌡️ Temp:        {temp}ºC")

            if alerta:
                print("🚨 ALERTA: ACIONANDO VENTILAÇÃO!")
            else:
                print("✅ Status: Normal")

    # --- TRATAMENTO DE ERROS (ROBUSTEZ) ---
    except OSError as erro_hw:
        print(f"⚠️ Erro de Hardware: {erro_hw}")
        print("Tentando novamente...")

    except KeyboardInterrupt:
        print("\n🛑 Sistema parado pelo usuário.")
        break
    except Exception as e:
        print(f"❌ Erro crítico: {e}")
    # Pausa para não travar o processador
    time.sleep(config["intervalo_leitura"])
```
