```python
print(f"exemplo2")
# Simulando um sistema de senha
senha_correta = "1234"
senha_digitada = input("Digite a senha de acesso: ")
if senha_digitada == senha_correta:
 # Este bloco só roda se a senha for igual
 print("✅ Acesso Liberado. Bem-vindo!")
 bloqueio = False
else:
 # Este bloco roda em qualquer outro caso
 print("❌ Acesso Negado. Senha incorreta.")
 bloqueio = True
print(f"Estado do bloqueio: {bloqueio}")

print(f"exemplo3")
nivel_bateria = int(input("NIVEL DA BATERIA(0 - 100):"))

if nivel_bateria >=80:
  print("STATUS: BATERIA CHEIA (VERDE)")
elif nivel_bateria >30:
  print("STATUS: OPERAÇÃO NORMAL (AMARELO)")
elif nivel_bateria >0:
  print("STATUS: BATERIA CRÍTICA (VERMELHO)")
else:
  print("STATUS: SEM CARGA")
```
