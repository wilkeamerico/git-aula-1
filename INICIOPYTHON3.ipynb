```python
leitura_serial = "25.5" # Perceba as aspas: isso é um TEXTO, não um número.
# Se tentássemos fazer isto, daria erro:
# print(leitura_serial + 1)
# Solução: Convertendo (Casting) para número Real (float)
temperatura = float(leitura_serial)
# Agora podemos fazer matemática
print(f"Temperatura calculada: {temperatura + 1.0}")
# O inverso também acontece: números precisam virar texto para serem concatenados
pino_str = str(2)
print("Pino selecionado: " + pino_str)
```
