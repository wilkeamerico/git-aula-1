```python
a = 10
b = 3
total_leituras = 50
tamanho_pacote = 8
#================================================================
print(f"Soma (+): {a + b}")
print(f"Subtração (-): {a - b}")
print(f"Multiplicação (*): {a * b}")
print(f"Divisão Real (/): {a / b}") # Resultado sempre float
print(f"Potência (**): {a ** b}") # 10 elevado a 3

#===============================================================

pacotes_completos = total_leituras // tamanho_pacote # Quantos grupos cheios?
sobra = total_leituras % tamanho_pacote # Quantos ficaram de fora?
print(f"Enviando {pacotes_completos} pacotes cheios.")
print(f"Sobraram {sobra} leituras para o próximo envio.")
```
